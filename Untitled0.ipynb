```python
# ==============================================================
# PCB Comparison â€” Auto-crop, Component Detection, OCR, JSON Diff,
# Red Box Visualization, and Gemini VLM Integration
# ==============================================================

!apt-get update -qq
!apt-get install -qq -y tesseract-ocr
!pip install -q pytesseract scikit-image google-generativeai

import os, io, json, base64, math
import cv2, numpy as np, pytesseract
from google.colab import files, drive
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
from getpass import getpass
import google.generativeai as genai

# ---------------- Config ----------------
OUT_DIR = "pcb_compare_out"
os.makedirs(OUT_DIR, exist_ok=True)
IMG_SIZE = 256

# ---------------- Gemini API ----------------
if "GOOGLE_API_KEY" not in os.environ or not os.environ["GOOGLE_API_KEY"]:
    os.environ["GOOGLE_API_KEY"] = getpass("ðŸ”‘ Enter your Gemini API key: ")

genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

# ---------------- Utility Functions ----------------
def encode_image_to_b64(path):
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

def auto_crop_board(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_green, upper_green = np.array([25, 30, 30]), np.array([100, 255, 255])
    mask = cv2.inRange(hsv, lower_green, upper_green)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours: return image
    x,y,w,h = cv2.boundingRect(max(contours, key=cv2.contourArea))
    return image[y:y+h, x:x+w]

def extract_text(roi):
    if roi is None or roi.size == 0: return "unlabeled"
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    _, th = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    txt = pytesseract.image_to_string(th, config="--psm 6").strip().replace("\n", " ")
    return txt if txt else "unlabeled"

def detect_components(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    edges = cv2.Canny(blur, 50, 150)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    comps = []
    for i,c in enumerate(contours):
        x,y,w,h = cv2.boundingRect(c)
        if w*h < 400 or w/h > 10 or h/w > 10: continue
        roi = img[y:y+h, x:x+w]
        label = extract_text(roi)
        comps.append({"label": label, "position": [int(x), int(y), int(w), int(h)]})
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 1)
        cv2.putText(img, label, (x,y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,255), 1)
    return img, comps

def compare_json(master, test):
    diffs = []
    for m in master["components"]:
        match = next((t for t in test["components"] if t["label"] == m["label"]), None)
        if not match:
            diffs.append({"label": m["label"], "status": "missing in test"})
        else:
            dx = abs(m["position"][0] - match["position"][0])
            dy = abs(m["position"][1] - match["position"][1])
            if dx > 10 or dy > 10:
                diffs.append({
                    "label": m["label"], "status": "moved",
                    "old_position": m["position"], "new_position": match["position"]
                })
    return {"differences": diffs}

# ---------------- File Upload ----------------
print("ðŸ“¤ Upload master and test PCB images")
uploaded = files.upload()
paths = list(uploaded.keys())
if len(paths) < 2:
    raise ValueError("Please upload both master and test images.")

master_img = cv2.imread(paths[0])
test_img   = cv2.imread(paths[1])

# ---------------- Processing ----------------
master_crop = auto_crop_board(master_img)
test_crop   = auto_crop_board(test_img)

master_detected, master_comps = detect_components(master_crop.copy())
test_detected, test_comps = detect_components(test_crop.copy())

master_json = {"components": master_comps}
test_json   = {"components": test_comps}

cv2.imwrite(f"{OUT_DIR}/master_detected.png", master_detected)
cv2.imwrite(f"{OUT_DIR}/test_detected.png", test_detected)

with open(f"{OUT_DIR}/master.json", "w") as f: json.dump(master_json, f, indent=2)
with open(f"{OUT_DIR}/test.json", "w") as f: json.dump(test_json, f, indent=2)

# ---------------- Compare JSONs ----------------
diff_json = compare_json(master_json, test_json)
with open(f"{OUT_DIR}/differences.json", "w") as f: json.dump(diff_json, f, indent=2)
print("âœ… Differences JSON saved.")

# ---------------- Visualize Differences ----------------
for d in diff_json["differences"]:
    if d["status"] == "missing in test":
        x,y,w,h = d["old_position"] if "old_position" in d else [0,0,0,0]
        cv2.rectangle(test_detected, (x,y), (x+w, y+h), (0,0,255), 2)
        cv2.putText(test_detected, f"{d['label']} missing", (x,y-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
    elif d["status"] == "moved":
        ox,oy,ow,oh = d["old_position"]
        nx,ny,nw,nh = d["new_position"]
        cv2.rectangle(test_detected, (nx,ny), (nx+nw,ny+nh), (0,0,255), 2)
        cv2.putText(test_detected, f"{d['label']} moved", (nx,ny-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)

cv2.imwrite(f"{OUT_DIR}/differences_visual.png", test_detected)
print("ðŸŸ¥ Differences visual saved.")

plt.imshow(cv2.cvtColor(test_detected, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# ---------------- Gemini VLM Payload ----------------
master_b64 = encode_image_to_b64(f"{OUT_DIR}/master_detected.png")
test_b64   = encode_image_to_b64(f"{OUT_DIR}/test_detected.png")

prompt = """
You are a Vision-Language Model specialized in PCB inspection.

Given two images (master PCB and test PCB):
1ï¸âƒ£ Detect and label components (ICs, resistors, capacitors, diodes, etc.).
2ï¸âƒ£ Read any visible text (e.g., R1, C2, IC5) using OCR.
3ï¸âƒ£ Generate structured JSON for each image:
   {"components": [{"label": "R1", "position": [x,y,w,h]}, ...]}
4ï¸âƒ£ Compare the two boards:
   - Identify missing, extra, or shifted components.
   - Include positional differences.
5ï¸âƒ£ Return a JSON of differences:
   {"differences": [{"label": "R1", "status": "missing in test"}, ...]}
6ï¸âƒ£ Be detailed and accurate.
Only output valid JSON â€” no explanation.
"""

vlm_payload = {
    "input": {
        "prompt": prompt,
        "master_image_b64": master_b64,
        "test_image_b64": test_b64
    }
}

with open(f"{OUT_DIR}/vlm_payload.json", "w") as f: json.dump(vlm_payload, f, indent=2)
print("âœ… VLM payload ready.")

# ---------------- Gemini VLM Call ----------------
run_vlm = input("Run Gemini VLM for visual comparison? (y/n): ").strip().lower()
if run_vlm == "y":
    try:
        response = genai.generate_content(
            model="gemini-1.5-pro",
            contents=[
                {"type": "text", "text": prompt},
                {"type": "image", "image_base64": master_b64},
                {"type": "image", "image_base64": test_b64},
            ]
        )

        print("\nðŸ” Gemini Response:")
        print(response.text)

        # Try parsing JSON
        try:
            start, end = response.text.find("{"), response.text.rfind("}") + 1
            result_json = json.loads(response.text[start:end])
            with open(f"{OUT_DIR}/vlm_result.json", "w") as f:
                json.dump(result_json, f, indent=2)
            print("\nðŸ’¾ VLM JSON saved as vlm_result.json")
        except:
            print("âš ï¸ Could not extract valid JSON â€” check raw text above.")

    except Exception as e:
        print("âŒ Gemini VLM call failed:", e)
else:
    print("Skipped Gemini VLM call. Payload saved for later use.")

print("\nâœ… All outputs saved in:", OUT_DIR)
```
