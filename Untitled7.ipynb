```python
# -------------------------------
# Step 1: Imports
# -------------------------------
import torch
import torch.nn as nn
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# -------------------------------
# Step 2: Define Siamese U-Net
# -------------------------------
class DoubleConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )
    def forward(self, x):
        return self.block(x)

class Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = DoubleConv(3,64)
        self.pool1 = nn.MaxPool2d(2)
        self.layer2 = DoubleConv(64,128)
        self.pool2 = nn.MaxPool2d(2)
        self.layer3 = DoubleConv(128,256)
        self.pool3 = nn.MaxPool2d(2)
        self.layer4 = DoubleConv(256,512)
    def forward(self, x):
        x1 = self.layer1(x)
        x2 = self.layer2(self.pool1(x1))
        x3 = self.layer3(self.pool2(x2))
        x4 = self.layer4(self.pool3(x3))
        return [x1, x2, x3, x4]

class Decoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.up3 = nn.ConvTranspose2d(512,256,2,stride=2)
        self.conv3 = DoubleConv(512,256)
        self.up2 = nn.ConvTranspose2d(256,128,2,stride=2)
        self.conv2 = DoubleConv(256,128)
        self.up1 = nn.ConvTranspose2d(128,64,2,stride=2)
        self.conv1 = DoubleConv(128,64)
        self.out = nn.Conv2d(64,1,1)
    def forward(self, x4, skips):
        x3, x2, x1 = skips[2], skips[1], skips[0]
        d3 = self.up3(x4)
        d3 = self.conv3(torch.cat([d3,x3], dim=1))
        d2 = self.up2(d3)
        d2 = self.conv2(torch.cat([d2,x2], dim=1))
        d1 = self.up1(d2)
        d1 = self.conv1(torch.cat([d1,x1], dim=1))
        return torch.sigmoid(self.out(d1))

class SiameseUNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = Encoder()
        self.decoder = Decoder()
    def forward(self, master, test):
        feats_m = self.encoder(master)
        feats_t = self.encoder(test)
        diff = [torch.abs(fm - ft) for fm, ft in zip(feats_m, feats_t)]
        mask = self.decoder(diff[-1], diff)
        return mask

# -------------------------------
# Step 3: Upload Master and Test Images
# -------------------------------
uploaded = files.upload()  # upload both master and test PCB

if len(uploaded.keys()) < 2:
    raise ValueError("Upload at least two images: master and test PCB.")

master_path = list(uploaded.keys())[0]
test_path   = list(uploaded.keys())[1]

master_img = cv2.imread(master_path)
test_img   = cv2.imread(test_path)

master_resized = cv2.resize(master_img, (512,512))
test_resized   = cv2.resize(test_img, (512,512))

master_tensor = torch.tensor(master_resized.transpose(2,0,1)/255.0, dtype=torch.float32).unsqueeze(0).to(device)
test_tensor   = torch.tensor(test_resized.transpose(2,0,1)/255.0, dtype=torch.float32).unsqueeze(0).to(device)

# -------------------------------
# Step 4: Initialize model (no training)
# -------------------------------
model = SiameseUNet().to(device)
model.eval()  # inference mode

# -------------------------------
# Step 5: Predict defect mask
# -------------------------------
with torch.no_grad():
    mask_pred = model(master_tensor, test_tensor)[0,0].cpu().numpy()

# Resize mask to original test image size
mask_resized = cv2.resize(mask_pred, (test_img.shape[1], test_img.shape[0]))

# -------------------------------
# Step 6: Visualize defect mask
# -------------------------------
plt.figure(figsize=(8,8))
plt.imshow(mask_resized, cmap='gray')
plt.title("Predicted Defect Mask")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import os
from google.colab import files
import matplotlib.pyplot as plt

# ==== Step 1: Upload images ====
print("Upload the MASTER PCB image first, then the TEST PCB image.")
uploaded = files.upload()

if len(uploaded.keys()) < 2:
    raise ValueError("Please upload at least two images: master and test PCB.")

master_path = list(uploaded.keys())[0]
test_path   = list(uploaded.keys())[1]

# ==== Step 2: Load images ====
master = cv2.imread(master_path)
test = cv2.imread(test_path)

# ==== Step 3: Detect and crop PCB function ====
def detect_and_crop(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    # Use adaptive threshold to handle lighting
    thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY_INV, 11, 2)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        # fallback: resize full image
        return cv2.resize(img, (512,512))

    # Assume largest contour is PCB
    c = max(contours, key=cv2.contourArea)
    x, y, w, h = cv2.boundingRect(c)
    cropped = img[y:y+h, x:x+w]
    resized = cv2.resize(cropped, (512,512))
    return resized

# ==== Step 4: Crop & resize master and test ====
master_cropped = detect_and_crop(master)
test_cropped   = detect_and_crop(test)

# ==== Step 5: Preprocess for difference detection ====
def preprocess(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    return blur

gray_master = preprocess(master_cropped)
gray_test   = preprocess(test_cropped)

# ==== Step 6: Compute absolute difference ====
diff = cv2.absdiff(gray_master, gray_test)

# ==== Step 7: Threshold & clean noise ====
thresh = cv2.adaptiveThreshold(diff, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                               cv2.THRESH_BINARY, 11, 2)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)

# ==== Step 8: Find contours & mark defects ====
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
boxed_img = test_cropped.copy()

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(boxed_img, (x, y), (x+w, y+h), (0,0,255), 2)  # red boxes

# ==== Step 9: Display results ====
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(master_cropped, cv2.COLOR_BGR2RGB))
plt.title("Cropped Master PCB")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(boxed_img, cv2.COLOR_BGR2RGB))
plt.title("Test PCB with Defects")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(thresh, cmap='gray')
plt.title("Difference Mask")
plt.axis('off')

plt.show()

# ==== Step 10: Save outputs ====
os.makedirs("output", exist_ok=True)
cv2.imwrite("output/cropped_master.png", master_cropped)
cv2.imwrite("output/cropped_test.png", test_cropped)
cv2.imwrite("output/test_with_defects.png", boxed_img)
cv2.imwrite("output/difference_mask.png", thresh)

print(" Completed. Outputs saved in 'output/' folder.")
```

```python
# Step 1: Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import os

# Step 2: Upload images
print("Upload MASTER PCB first, then TEST PCB")
uploaded = files.upload()

if len(uploaded.keys()) < 2:
    raise ValueError("Upload at least two images: master and test PCB.")

master_path = list(uploaded.keys())[0]
test_path = list(uploaded.keys())[1]

master_img = cv2.imread(master_path)
test_img = cv2.imread(test_path)

# Step 3: ECC-based alignment on full images
def align_ecc(master, test):
    gray_master = cv2.cvtColor(master, cv2.COLOR_BGR2GRAY)
    gray_test = cv2.cvtColor(test, cv2.COLOR_BGR2GRAY)

    warp_matrix = np.eye(2, 3, dtype=np.float32)  # affine
    criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 5000, 1e-10)

    try:
        cc, warp_matrix = cv2.findTransformECC(gray_master, gray_test, warp_matrix, cv2.MOTION_AFFINE, criteria)
        aligned = cv2.warpAffine(test, warp_matrix, (master.shape[1], master.shape[0]),
                                 flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
    except:
        aligned = cv2.resize(test, (master.shape[1], master.shape[0]))
    return aligned

test_aligned = align_ecc(master_img, test_img)

# Step 4: Optional minimal cropping (to remove borders)
def minimal_crop(img, margin=10):
    return img[margin:-margin, margin:-margin]

master_cropped = minimal_crop(master_img, margin=10)
test_cropped = minimal_crop(test_aligned, margin=10)

# Step 5: Pixel-wise difference
gray_master = cv2.cvtColor(master_cropped, cv2.COLOR_BGR2GRAY)
gray_test = cv2.cvtColor(test_cropped, cv2.COLOR_BGR2GRAY)
gray_master = cv2.GaussianBlur(gray_master, (5,5), 0)
gray_test = cv2.GaussianBlur(gray_test, (5,5), 0)

diff = cv2.absdiff(gray_master, gray_test)

_, thresh = cv2.threshold(diff, 50, 255, cv2.THRESH_BINARY)
kernel = np.ones((3,3), np.uint8)
thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)

# Step 6: Find defects
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
boxed_img = test_cropped.copy()
for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    if w*h > 50:
        cv2.rectangle(boxed_img, (x,y), (x+w, y+h), (0,0,255), 2)

# Step 7: Display
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(master_cropped, cv2.COLOR_BGR2RGB))
plt.title("Master PCB"); plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(test_cropped, cv2.COLOR_BGR2RGB))
plt.title("Test PCB Aligned"); plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(boxed_img[...,::-1])
plt.title("Detected Defects"); plt.axis('off')

plt.show()

# Step 8: Save
os.makedirs("output", exist_ok=True)
cv2.imwrite("output/master.png", master_cropped)
cv2.imwrite("output/test_aligned.png", test_cropped)
cv2.imwrite("output/difference_mask.png", thresh)
cv2.imwrite("output/test_with_defects.png", boxed_img)

print(" Completed. Outputs saved in 'output/' folder.")
```

```python
# -------------------------------
# Step 1: Imports
# -------------------------------
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -------------------------------
# Step 2: Upload image
# -------------------------------
print(" Upload the PCB image you want to crop:")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
print(f" Image uploaded: {filename}")

# -------------------------------
# Step 3: Improved PCB Cropping Function
# -------------------------------
def auto_crop_board_color_based(image):
    img = image.copy()
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    #  Define color range for PCB (tuned for green/teal boards)
    lower_green = np.array([25, 30, 30])
    upper_green = np.array([100, 255, 255])

    #  Create color mask
    mask = cv2.inRange(hsv, lower_green, upper_green)

    #  Morphological cleanup
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)

    #  Find contours from mask
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        print(" No PCB region detected via color  fallback to full image.")
        return cv2.resize(img, (512, 512))

    #  Select largest contour (most likely the PCB)
    c = max(contours, key=cv2.contourArea)

    # Approximate to get corners
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)

    if len(approx) != 4:
        # If not perfect 4-corner, get bounding rectangle
        x, y, w, h = cv2.boundingRect(c)
        cropped = img[y:y + h, x:x + w]
        return cropped

    #  Order corner points
    def order_points(pts):
        rect = np.zeros((4, 2), dtype="float32")
        s = pts.sum(axis=1)
        rect[0] = pts[np.argmin(s)]      # top-left
        rect[2] = pts[np.argmax(s)]      # bottom-right
        diff = np.diff(pts, axis=1)
        rect[1] = pts[np.argmin(diff)]   # top-right
        rect[3] = pts[np.argmax(diff)]   # bottom-left
        return rect

    rect = order_points(approx.reshape(4, 2))
    (tl, tr, br, bl) = rect

    # Compute new perspective size
    widthA = np.linalg.norm(br - bl)
    widthB = np.linalg.norm(tr - tl)
    heightA = np.linalg.norm(tr - br)
    heightB = np.linalg.norm(tl - bl)
    maxWidth = int(max(widthA, widthB))
    maxHeight = int(max(heightA, heightB))

    # Warp perspective (get perfect top-down PCB)
    dst = np.array([
        [0, 0],
        [maxWidth - 1, 0],
        [maxWidth - 1, maxHeight - 1],
        [0, maxHeight - 1]], dtype="float32")

    M = cv2.getPerspectiveTransform(rect, dst)
    warped = cv2.warpPerspective(img, M, (maxWidth, maxHeight))

    return warped

# -------------------------------
# Step 4: Read, crop, and show result
# -------------------------------
img = cv2.imread(filename)
cropped = auto_crop_board_color_based(img)

# Convert BGR  RGB for display
cropped_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(cropped_rgb)
plt.axis('off')
plt.title(" Perfectly Cropped PCB Region")
plt.show()

# -------------------------------
# Step 5: Save Output
# -------------------------------
output_name = "cropped_pcb_clean.jpg"
cv2.imwrite(output_name, cropped)
print(f" Cropped image saved as: {output_name}")
files.download(output_name)
```

```python
# -------------------------------
# Step 1: Imports
# -------------------------------
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import os

# -------------------------------
# Step 2: Upload images
# -------------------------------
print(" Upload MASTER PCB first, then TEST PCB")
uploaded = files.upload()

if len(uploaded.keys()) < 2:
    raise ValueError("Upload at least two images: master and test PCB.")

master_path = list(uploaded.keys())[0]
test_path   = list(uploaded.keys())[1]

master_img = cv2.imread(master_path)
test_img   = cv2.imread(test_path)

# -------------------------------
# Step 3: PCB Cropping Function (Color-based)
# -------------------------------
def auto_crop_board_color_based(image):
    img = image.copy()
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    #  Define color range for PCB (tuned for green/teal boards)
    lower_green = np.array([25, 30, 30])
    upper_green = np.array([100, 255, 255])

    #  Create color mask
    mask = cv2.inRange(hsv, lower_green, upper_green)

    #  Morphological cleanup
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)

    #  Find contours from mask
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        print(" No PCB region detected via color  fallback to full image.")
        return cv2.resize(img, (512, 512))

    #  Select largest contour (most likely the PCB)
    c = max(contours, key=cv2.contourArea)

    # Approximate to get corners
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)

    if len(approx) != 4:
        # If not perfect 4-corner, get bounding rectangle
        x, y, w, h = cv2.boundingRect(c)
        cropped = img[y:y + h, x:x + w]
        return cropped

    #  Order corner points
    def order_points(pts):
        rect = np.zeros((4, 2), dtype="float32")
        s = pts.sum(axis=1)
        rect[0] = pts[np.argmin(s)]      # top-left
        rect[2] = pts[np.argmax(s)]      # bottom-right
        diff = np.diff(pts, axis=1)
        rect[1] = pts[np.argmin(diff)]   # top-right
        rect[3] = pts[np.argmax(diff)]   # bottom-left
        return rect

    rect = order_points(approx.reshape(4, 2))
    (tl, tr, br, bl) = rect

    # Compute new perspective size
    widthA = np.linalg.norm(br - bl)
    widthB = np.linalg.norm(tr - tl)
    heightA = np.linalg.norm(tr - br)
    heightB = np.linalg.norm(tl - bl)
    maxWidth = int(max(widthA, widthB))
    maxHeight = int(max(heightA, heightB))

    # Warp perspective (get perfect top-down PCB)
    dst = np.array([
        [0, 0],
        [maxWidth - 1, 0],
        [maxWidth - 1, maxHeight - 1],
        [0, maxHeight - 1]], dtype="float32")

    M = cv2.getPerspectiveTransform(rect, dst)
    warped = cv2.warpPerspective(img, M, (maxWidth, maxHeight))

    return warped

# -------------------------------
# Step 4: Crop both Master and Test boards
# -------------------------------
master_cropped = auto_crop_board_color_based(master_img)
test_cropped   = auto_crop_board_color_based(test_img)

# -------------------------------
# Step 5: ECC-based alignment
# -------------------------------
def align_ecc(master, test):
    gray_master = cv2.cvtColor(master, cv2.COLOR_BGR2GRAY)
    gray_test   = cv2.cvtColor(test, cv2.COLOR_BGR2GRAY)

    warp_matrix = np.eye(2, 3, dtype=np.float32)  # affine
    criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 5000, 1e-10)

    try:
        cc, warp_matrix = cv2.findTransformECC(gray_master, gray_test, warp_matrix, cv2.MOTION_AFFINE, criteria)
        aligned = cv2.warpAffine(test, warp_matrix, (master.shape[1], master.shape[0]),
                                 flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
    except:
        aligned = cv2.resize(test, (master.shape[1], master.shape[0]))
    return aligned

test_aligned = align_ecc(master_cropped, test_cropped)

# -------------------------------
# Step 6: Optional minimal cropping (border cleanup)
# -------------------------------
def minimal_crop(img, margin=10):
    return img[margin:-margin, margin:-margin]

master_cropped_final = minimal_crop(master_cropped, margin=10)
test_cropped_final   = minimal_crop(test_aligned, margin=10)

# -------------------------------
# Step 7: Pixel-wise difference & defect marking
# -------------------------------
gray_master = cv2.cvtColor(master_cropped_final, cv2.COLOR_BGR2GRAY)
gray_test   = cv2.cvtColor(test_cropped_final, cv2.COLOR_BGR2GRAY)
gray_master = cv2.GaussianBlur(gray_master, (5,5), 0)
gray_test   = cv2.GaussianBlur(gray_test, (5,5), 0)

diff = cv2.absdiff(gray_master, gray_test)

_, thresh = cv2.threshold(diff, 50, 255, cv2.THRESH_BINARY)
kernel = np.ones((3,3), np.uint8)
thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)

# Find contours (defects)
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
boxed_img = test_cropped_final.copy()
for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    if w*h > 50:
        cv2.rectangle(boxed_img, (x,y), (x+w, y+h), (0,0,255), 2)

# -------------------------------
# Step 8: Display Results
# -------------------------------
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(master_cropped_final, cv2.COLOR_BGR2RGB))
plt.title("Master PCB"); plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(test_cropped_final, cv2.COLOR_BGR2RGB))
plt.title("Test PCB Aligned"); plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(boxed_img[...,::-1])
plt.title("Detected Defects"); plt.axis('off')

plt.show()

# -------------------------------
# Step 9: Save Outputs
# -------------------------------
os.makedirs("output", exist_ok=True)
cv2.imwrite("output/master.png", master_cropped_final)
cv2.imwrite("output/test_aligned.png", test_cropped_final)
cv2.imwrite("output/difference_mask.png", thresh)
cv2.imwrite("output/test_with_defects.png", boxed_img)

print(" Completed. Outputs saved in 'output/' folder.")
```
