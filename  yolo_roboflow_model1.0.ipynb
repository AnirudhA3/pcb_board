```python
import os, random, shutil
from pathlib import Path

base = Path("/content/drive/My Drive/YOLO/pcb_dataset")
images = list((base / "images").glob("*.jpg"))
random.shuffle(images)

split_ratio = 0.8
train_count = int(len(images) * split_ratio)

train_imgs = images[:train_count]
val_imgs = images[train_count:]

def move_files(file_list, subset):
    (base / subset / "images").mkdir(parents=True, exist_ok=True)
    (base / subset / "labels").mkdir(parents=True, exist_ok=True)
    for img_path in file_list:
        label_path = base / "labels" / (img_path.stem + ".txt")
        if label_path.exists():
            shutil.copy(img_path, base / subset / "images" / img_path.name)
            shutil.copy(label_path, base / subset / "labels" / label_path.name)

move_files(train_imgs, "train")
move_files(val_imgs, "valid")

print(f"âœ… Split complete: {len(train_imgs)} train, {len(val_imgs)} valid images.")
```

```python
# ==============================================================
# PCB COMPARISON â€” YOLOv8 | JSON Diff | Red Box Visualization
# ==============================================================

# ðŸ§° Requirements
!pip install -q ultralytics opencv-python Pillow matplotlib numpy

# ==============================================================
# ðŸ§  Imports
# ==============================================================
import os, cv2, json
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from google.colab import files

# ==============================================================
# ðŸ”§ Setup
# ==============================================================
OUT_DIR = "/content/pcb_output"
os.makedirs(OUT_DIR, exist_ok=True)

# ðŸ‘‡ Path to your trained YOLOv8 model
MODEL_PATH = "/content/drive/MyDrive/YOLO/runs/pcb_yolov8_cap_res/weights/best.pt"

model = YOLO(MODEL_PATH)
print("âœ… YOLOv8 model loaded successfully from:", MODEL_PATH)

# ==============================================================
# ðŸ§© Utility Functions
# ==============================================================
def auto_crop_board(image):
    """Auto-crops the PCB area using green color segmentation."""
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_green = np.array([25, 30, 30])
    upper_green = np.array([100, 255, 255])
    mask = cv2.inRange(hsv, lower_green, upper_green)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return image
    x, y, w, h = cv2.boundingRect(max(contours, key=cv2.contourArea))
    return image[y:y+h, x:x+w]

def detect_with_yolo(image, prefix="B"):
    """Runs YOLO detection on a PCB image and returns annotated image + JSON."""
    results = model(image)
    annotated = results[0].plot()  # image with boxes
    comps = []
    for i, box in enumerate(results[0].boxes):
        cls_id = int(box.cls)
        label = results[0].names[cls_id]
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
        comps.append({
            "id": f"{prefix}{i}",
            "label": label,
            "position": [int(x1), int(y1), int(x2 - x1), int(y2 - y1)]
        })
    return annotated, comps

def compare_json(master, test):
    """Compares master and test JSONs for missing or moved components."""
    diffs = []
    for m in master["components"]:
        match = next((t for t in test["components"] if t["label"] == m["label"]), None)
        if not match:
            diffs.append({"label": m["label"], "status": "missing in test", "old_position": m["position"]})
        else:
            dx = abs(m["position"][0] - match["position"][0])
            dy = abs(m["position"][1] - match["position"][1])
            if dx > 10 or dy > 10:
                diffs.append({
                    "label": m["label"], "status": "moved",
                    "old_position": m["position"],
                    "new_position": match["position"]
                })
    return {"differences": diffs}

def visualize_differences(diff_json, test_img):
    """Draws red boxes on moved/missing components."""
    for d in diff_json["differences"]:
        if d["status"] == "missing in test":
            x, y, w, h = d["old_position"]
            cv2.rectangle(test_img, (x,y), (x+w,y+h), (0,0,255), 2)
            cv2.putText(test_img, f"{d['label']} missing", (x, y-5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
        elif d["status"] == "moved":
            nx, ny, nw, nh = d["new_position"]
            cv2.rectangle(test_img, (nx,ny), (nx+nw,ny+nh), (0,0,255), 2)
            cv2.putText(test_img, f"{d['label']} moved", (nx,ny-5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
    return test_img

# ==============================================================
# ðŸ“¤ Upload PCB Images
# ==============================================================
print("ðŸ“¸ Upload your MASTER and TEST PCB images")
uploaded = files.upload()
paths = list(uploaded.keys())
if len(paths) < 2:
    raise ValueError("Please upload both master and test PCB images.")

master_img = cv2.imread(paths[0])
test_img   = cv2.imread(paths[1])

# ==============================================================
# ðŸ” YOLO Detection + JSON Save
# ==============================================================
master_crop = auto_crop_board(master_img)
test_crop   = auto_crop_board(test_img)

master_detected, master_comps = detect_with_yolo(master_crop, "M")
test_detected, test_comps = detect_with_yolo(test_crop, "T")

master_json = {"components": master_comps}
test_json   = {"components": test_comps}

cv2.imwrite(f"{OUT_DIR}/master_detected.png", master_detected)
cv2.imwrite(f"{OUT_DIR}/test_detected.png", test_detected)
with open(f"{OUT_DIR}/master.json", "w") as f: json.dump(master_json, f, indent=2)
with open(f"{OUT_DIR}/test.json", "w") as f: json.dump(test_json, f, indent=2)

print("âœ… YOLO detection complete. JSON files saved to:", OUT_DIR)

# ==============================================================
# ðŸ” Compare JSONs + Visualize Differences
# ==============================================================
diff_json = compare_json(master_json, test_json)
with open(f"{OUT_DIR}/differences.json", "w") as f: json.dump(diff_json, f, indent=2)

visualized = visualize_differences(diff_json, test_detected.copy())
cv2.imwrite(f"{OUT_DIR}/difference_visual.png", visualized)

plt.imshow(cv2.cvtColor(visualized, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

print("ðŸŸ¥ Visual comparison saved at:", f"{OUT_DIR}/difference_visual.png")
```

```python
# ==============================================================
# PCB COMPARISON â€” YOLO + OCR + CAE + JSON Diff + Visualization
# ==============================================================

# ðŸ§° Requirements
!apt-get update -qq
!apt-get install -qq -y tesseract-ocr
!pip install -q ultralytics opencv-python pytesseract scikit-image numpy Pillow matplotlib

# ==============================================================
# ðŸ§  Imports
# ==============================================================
import os, cv2, json, pytesseract
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from google.colab import files

# ==============================================================
# ðŸ”§ Setup
# ==============================================================
OUT_DIR = "/content/pcb_output"
os.makedirs(OUT_DIR, exist_ok=True)

# ðŸ‘‡ Your trained YOLO model
MODEL_PATH = "/content/drive/MyDrive/YOLO/runs/pcb_yolov8_cap_res/weights/best.pt"
model = YOLO(MODEL_PATH)
print("âœ… YOLOv8 model loaded successfully from:", MODEL_PATH)

# ==============================================================
# ðŸ§© Utility Functions
# ==============================================================
def auto_crop_board(image):
    """Auto-crops the PCB area using green color segmentation."""
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_green = np.array([25, 30, 30])
    upper_green = np.array([100, 255, 255])
    mask = cv2.inRange(hsv, lower_green, upper_green)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return image
    x, y, w, h = cv2.boundingRect(max(contours, key=cv2.contourArea))
    return image[y:y+h, x:x+w]

def extract_text(roi):
    """OCR: Extract component label text."""
    if roi is None or roi.size == 0:
        return "unlabeled"
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    _, th = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    txt = pytesseract.image_to_string(th, config="--psm 6").strip().replace("\n", " ")
    return txt if txt else "unlabeled"

# ==============================================================
# ðŸ¤– YOLO + OCR + CAE Detection
# ==============================================================
def detect_with_yolo_ocr_cae(image, prefix="B"):
    """
    Runs YOLO detection + OCR for text labels.
    CAE refines the results (basic rule-based verification for now).
    """
    results = model(image)
    annotated = results[0].plot()  # Draw YOLO boxes

    comps = []
    for i, box in enumerate(results[0].boxes):
        cls_id = int(box.cls)
        label = results[0].names[cls_id]
        conf = float(box.conf.cpu().numpy())
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)

        roi = image[y1:y2, x1:x2]
        ocr_text = extract_text(roi)

        # === CAE (Context-Aware Enhancer) ===
        # Idea: Verify if OCR text matches or supports YOLO label
        verified_label = label
        if ocr_text.lower() in label.lower() or label.lower() in ocr_text.lower():
            verified_label = f"{label} âœ…"
        elif conf < 0.5 and ocr_text != "unlabeled":
            verified_label = f"{label}? ({ocr_text})"  # uncertain prediction

        comps.append({
            "id": f"{prefix}{i}",
            "label": verified_label,
            "ocr": ocr_text,
            "confidence": round(conf, 3),
            "position": [int(x1), int(y1), int(x2 - x1), int(y2 - y1)]
        })

        # Draw annotated boxes
        cv2.rectangle(annotated, (x1, y1), (x2, y2), (0,255,0), 2)
        cv2.putText(annotated, verified_label, (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,255), 1)

    return annotated, comps

# ==============================================================
# ðŸ§® JSON Comparison
# ==============================================================
def compare_json(master, test):
    """Compares master and test JSONs for missing/moved components."""
    diffs = []
    for m in master["components"]:
        match = next((t for t in test["components"] if t["label"].split()[0] == m["label"].split()[0]), None)
        if not match:
            diffs.append({"label": m["label"], "status": "missing in test", "old_position": m["position"]})
        else:
            dx = abs(m["position"][0] - match["position"][0])
            dy = abs(m["position"][1] - match["position"][1])
            if dx > 10 or dy > 10:
                diffs.append({
                    "label": m["label"], "status": "moved",
                    "old_position": m["position"],
                    "new_position": match["position"]
                })
    return {"differences": diffs}

# ==============================================================
# ðŸŽ¨ Visualization
# ==============================================================
def visualize_differences(diff_json, test_img):
    """Draws red boxes on moved/missing components."""
    for d in diff_json["differences"]:
        if d["status"] == "missing in test":
            x,y,w,h = d["old_position"]
            cv2.rectangle(test_img, (x,y), (x+w,y+h), (0,0,255), 2)
            cv2.putText(test_img, f"{d['label']} missing", (x,y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
        elif d["status"] == "moved":
            nx,ny,nw,nh = d["new_position"]
            cv2.rectangle(test_img, (nx,ny), (nx+nw,ny+nh), (0,0,255), 2)
            cv2.putText(test_img, f"{d['label']} moved", (nx,ny-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
    return test_img

# ==============================================================
# ðŸ“¤ Upload PCB Images
# ==============================================================
print("ðŸ“¸ Upload your MASTER and TEST PCB images")
uploaded = files.upload()
paths = list(uploaded.keys())
if len(paths) < 2:
    raise ValueError("Please upload both master and test PCB images.")

master_img = cv2.imread(paths[0])
test_img   = cv2.imread(paths[1])

# ==============================================================
# ðŸ” Detection + JSON Save
# ==============================================================
master_crop = auto_crop_board(master_img)
test_crop   = auto_crop_board(test_img)

master_detected, master_comps = detect_with_yolo_ocr_cae(master_crop.copy(), "M")
test_detected, test_comps = detect_with_yolo_ocr_cae(test_crop.copy(), "T")

master_json = {"components": master_comps}
test_json   = {"components": test_comps}

cv2.imwrite(f"{OUT_DIR}/master_detected.png", master_detected)
cv2.imwrite(f"{OUT_DIR}/test_detected.png", test_detected)
with open(f"{OUT_DIR}/master.json", "w") as f: json.dump(master_json, f, indent=2)
with open(f"{OUT_DIR}/test.json", "w") as f: json.dump(test_json, f, indent=2)

print("âœ… YOLO + OCR detection complete. JSON files saved to:", OUT_DIR)

# ==============================================================
# ðŸ” Compare JSONs + Visualize
# ==============================================================
diff_json = compare_json(master_json, test_json)
with open(f"{OUT_DIR}/differences.json", "w") as f: json.dump(diff_json, f, indent=2)

visualized = visualize_differences(diff_json, test_detected.copy())
cv2.imwrite(f"{OUT_DIR}/difference_visual.png", visualized)

plt.imshow(cv2.cvtColor(visualized, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

print("ðŸŸ¥ Visual comparison saved at:", f"{OUT_DIR}/difference_visual.png")
```

```python
# ===============================================================
# CAE (Context-Aware Enhancer) â€” Co-Pilot for YOLO PCB Detection
# Clean + Stable Version for Colab
# ===============================================================

# ðŸ§° Dependencies
!apt-get update -qq
!apt-get install -qq -y tesseract-ocr
!pip install -q ultralytics torch torchvision pytesseract opencv-python matplotlib tqdm

# ===============================================================
# ðŸ§  Imports
# ===============================================================
import os, cv2, json, pytesseract
import numpy as np
from tqdm import tqdm
from pathlib import Path

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
from torchvision.models import mobilenet_v2, MobileNet_V2_Weights

from ultralytics import YOLO

# ===============================================================
# âš™ï¸ Setup
# ===============================================================
DRIVE_ROOT = "/content/drive/MyDrive/YOLO"
DATA_ROOT = f"{DRIVE_ROOT}/pcb_dataset"   # contains train/images and train/labels
YOLO_WEIGHTS = f"{DRIVE_ROOT}/runs/pcb_yolov8_cap_res/weights/best.pt"
CAE_SAVE = f"{DRIVE_ROOT}/cae_model.pth"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("âœ… Device:", DEVICE)

# ===============================================================
# ðŸ” Helper Functions
# ===============================================================
def load_yolo():
    model = YOLO(YOLO_WEIGHTS)
    model.to(DEVICE)
    return model

def read_yolo_label(txt_path):
    items = []
    if not os.path.exists(txt_path):
        return items
    with open(txt_path, "r") as f:
        for l in f:
            parts = l.strip().split()
            if len(parts) != 5:
                continue
            cls = int(parts[0])
            x, y, w, h = map(float, parts[1:])
            items.append((cls, x, y, w, h))
    return items

def xywh_norm_to_xyxy(img_w, img_h, x_c, y_c, w, h):
    x1 = int((x_c - w/2) * img_w)
    y1 = int((y_c - h/2) * img_h)
    x2 = int((x_c + w/2) * img_w)
    y2 = int((y_c + h/2) * img_h)
    return max(0,x1), max(0,y1), min(img_w-1,x2), min(img_h-1,y2)

def iou(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interW = max(0, xB-xA+1)
    interH = max(0, yB-yA+1)
    inter = interW * interH
    boxAA = (boxA[2]-boxA[0]+1)*(boxA[3]-boxA[1]+1)
    boxBB = (boxB[2]-boxB[0]+1)*(boxB[3]-boxB[1]+1)
    union = float(boxAA + boxBB - inter)
    return inter / union if union > 0 else 0.0

# ===============================================================
# ðŸ§© OCR for CAE
# ===============================================================
def extract_text_for_cae(roi):
    if roi is None or roi.size == 0:
        return "unlabeled"
    roi = cv2.resize(roi, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    gray = cv2.fastNlMeansDenoising(gray, h=30)
    th = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                               cv2.THRESH_BINARY, 31, 2)
    kernel = np.ones((2,2), np.uint8)
    th = cv2.morphologyEx(th, cv2.MORPH_OPEN, kernel)
    txt = pytesseract.image_to_string(
        th,
        config="--psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
    ).strip().replace("\n", " ")
    return txt if txt else "unlabeled"

# ===============================================================
# ðŸ§± Build CAE Dataset
# ===============================================================
def build_cae_dataset(yolo_model, data_root=DATA_ROOT, split="train", iou_thresh=0.5):
    samples = []
    image_files = []
    for ext in ["*.jpg", "*.jpeg", "*.png"]:
        image_files.extend((Path(data_root) / split / "images").glob(ext))
    image_files = sorted(image_files)
    print(f"ðŸ“‚ Found {len(image_files)} {split} images")

    for img_path in tqdm(image_files, desc=f"Building {split} CAE samples"):
        img = cv2.imread(str(img_path))
        if img is None:
            continue
        h, w = img.shape[:2]

        preds = yolo_model.predict(source=str(img_path), conf=0.05, save=False, verbose=False)
        r = preds[0]

        gt_labels = read_yolo_label(str(Path(data_root) / split / "labels" / (img_path.stem + ".txt")))
        gt_boxes = [{'cls': cls, 'xyxy': xywh_norm_to_xyxy(w,h,xc,yc,ww,hh)} for cls,xc,yc,ww,hh in gt_labels]

        for b in r.boxes:
            xyxy = b.xyxy[0].cpu().numpy().astype(int)
            conf = float(b.conf.cpu().numpy())
            cls = int(b.cls)
            x1,y1,x2,y2 = xyxy

            best_iou, best_gt = 0, None
            for g in gt_boxes:
                val = iou(xyxy, g['xyxy'])
                if val > best_iou:
                    best_iou, best_gt = val, g

            crop = img[max(0,y1):min(h,y2), max(0,x1):min(w,x2)].copy()
            txt = extract_text_for_cae(crop)

            if best_gt is None or best_iou < iou_thresh:
                target_keep, target_class = 0, -1
            else:
                target_class = best_gt['cls']
                target_keep = 1 if cls == best_gt['cls'] else 0

            samples.append({
                'crop': crop,
                'yolo_cls': cls,
                'yolo_conf': conf,
                'ocr_text': txt,
                'target_keep': target_keep,
                'target_class': target_class
            })
    return samples

# ===============================================================
# ðŸ§  CAE Model
# ===============================================================
class SimpleTextEncoder(nn.Module):
    def __init__(self, vocab, embed_dim=32):
        super().__init__()
        self.vocab = vocab
        self.char2idx = {c:i+1 for i,c in enumerate(vocab)}
        self.embedding = nn.Embedding(len(self.char2idx)+1, embed_dim, padding_idx=0)
    def forward(self, texts):
        max_len = max([len(t) for t in texts])
        batch = []
        for t in texts:
            ints = [self.char2idx.get(ch, 0) for ch in t.upper()[:max_len]]
            ints += [0] * (max_len - len(ints))
            batch.append(ints)
        x = torch.tensor(batch, dtype=torch.long, device=DEVICE)
        return self.embedding(x).mean(dim=1)

class CAEModel(nn.Module):
    def __init__(self, num_classes, text_embed_dim=32, hidden=512):
        super().__init__()
        m = mobilenet_v2(weights=MobileNet_V2_Weights.IMAGENET1K_V1)
        self.visual = nn.Sequential(*list(m.features.children()))
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        self.vis_fc = nn.Linear(1280, 256)
        self.text_encoder = SimpleTextEncoder(list("ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"), text_embed_dim)
        fusion_dim = 256 + text_embed_dim + 2
        self.mlp = nn.Sequential(
            nn.Linear(fusion_dim, hidden),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden, hidden//2),
            nn.ReLU()
        )
        self.keep_head = nn.Linear(hidden//2, 1)
        self.class_head = nn.Linear(hidden//2, num_classes)

    def forward(self, images, texts, confs):
        x = self.visual(images)
        x = self.global_pool(x).flatten(1)
        x = self.vis_fc(x)
        t = self.text_encoder(texts)
        fused = torch.cat([x, t, confs], dim=1)
        h = self.mlp(fused)
        keep_logit = self.keep_head(h).squeeze(1)
        class_logits = self.class_head(h)
        return keep_logit, class_logits

# ===============================================================
# ðŸ§° Dataset and Loader
# ===============================================================
class CAEDataset(Dataset):
    def __init__(self, samples, num_classes):
        self.samples = samples
        self.num_classes = num_classes
        self.tf = T.Compose([
            T.ToPILImage(),
            T.Resize((224,224)),
            T.ToTensor(),
            T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
        ])
    def __len__(self): return len(self.samples)
    def __getitem__(self, idx):
        s = self.samples[idx]
        img_t = self.tf(s['crop'])
        conf = torch.tensor([s['yolo_conf'], s['yolo_cls']/max(1,self.num_classes-1)], dtype=torch.float)
        keep = torch.tensor(float(s['target_keep']))
        tclass = torch.tensor(max(0, s['target_class']))
        return img_t, s['ocr_text'], conf, keep, tclass

def collate_fn(batch):
    imgs = torch.stack([b[0] for b in batch]).to(DEVICE)
    texts = [b[1] for b in batch]
    confs = torch.stack([b[2] for b in batch]).to(DEVICE)
    keeps = torch.stack([b[3] for b in batch]).to(DEVICE)
    classes = torch.stack([b[4] for b in batch]).to(DEVICE)
    return imgs, texts, confs, keeps, classes

# ===============================================================
# ðŸ‹ï¸â€â™‚ï¸ Training
# ===============================================================
def train_cae(model, dataloader, epochs=5, lr=1e-4):
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    bce = nn.BCEWithLogitsLoss()
    ce = nn.CrossEntropyLoss()
    model.to(DEVICE)

    for ep in range(epochs):
        model.train()
        total = 0
        for imgs, texts, confs, keeps, classes in tqdm(dataloader, desc=f"Epoch {ep+1}/{epochs}"):
            opt.zero_grad()
            k_logits, c_logits = model(imgs, texts, confs)
            loss = bce(k_logits, keeps) + 0.5 * ce(c_logits, classes)
            loss.backward()
            opt.step()
            total += loss.item()
        print(f"âœ… Epoch {ep+1}: avg loss {total/len(dataloader):.4f}")
    print("ðŸŽ¯ Training complete!")

# ===============================================================
# ðŸš€ Run Everything
# ===============================================================
yolo = load_yolo()
train_samples = build_cae_dataset(yolo, DATA_ROOT, "train", iou_thresh=0.5)
print(f"ðŸ§¾ Total training samples: {len(train_samples)}")

num_classes = 2  # adjust if you have more components
dataset = CAEDataset(train_samples, num_classes)
loader = DataLoader(dataset, batch_size=8, shuffle=True, collate_fn=collate_fn, num_workers=0)

cae = CAEModel(num_classes)
train_cae(cae, loader, epochs=5, lr=1e-4)
torch.save(cae.state_dict(), CAE_SAVE)
print(f"ðŸ’¾ CAE model saved at {CAE_SAVE}")
```

```python
# ==============================================================
# PCB DETECTION â€” YOLO + Enhanced OCR + Correct CAE Architecture (Single Image)
# ==============================================================

!apt-get update -qq
!apt-get install -qq -y tesseract-ocr
!pip install -q ultralytics opencv-python pytesseract torch torchvision numpy Pillow matplotlib

# ==============================================================
# ðŸ§  Imports
# ==============================================================
import os, cv2, pytesseract, json, torch, numpy as np, matplotlib.pyplot as plt
from ultralytics import YOLO
from torchvision import transforms
from torch import nn
from google.colab import files

# ==============================================================
# ðŸ”§ Setup
# ==============================================================
OUT_DIR = "/content/pcb_output"
os.makedirs(OUT_DIR, exist_ok=True)

YOLO_PATH = "/content/drive/MyDrive/YOLO/runs/pcb_yolov8_cap_res/weights/best.pt"
CAE_PATH = "/content/drive/MyDrive/YOLO/cae_model.pth"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", DEVICE)

# Load YOLO
model = YOLO(YOLO_PATH)
print("âœ… YOLOv8 model loaded successfully.")

# ==============================================================
# ðŸ§© Correct CAE Model Definition (matches your trained file)
# ==============================================================
class CAEModel(nn.Module):
    def __init__(self, num_classes=2):
        super(CAEModel, self).__init__()
        self.visual = nn.Sequential(
            nn.Sequential(
                nn.Conv2d(3, 32, 3, stride=1, padding=1),
                nn.BatchNorm2d(32),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(2),
            ),
            nn.Sequential(
                nn.Conv2d(32, 64, 3, stride=1, padding=1),
                nn.BatchNorm2d(64),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(2),
            ),
            nn.Sequential(
                nn.Conv2d(64, 128, 3, stride=1, padding=1),
                nn.BatchNorm2d(128),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(2),
            ),
            nn.Sequential(
                nn.Conv2d(128, 256, 3, stride=1, padding=1),
                nn.BatchNorm2d(256),
                nn.ReLU(inplace=True),
                nn.AdaptiveAvgPool2d((1,1)),
            ),
        )
        self.fc = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        feats = self.visual[0](x)
        for i in range(1, len(self.visual)):
            feats = self.visual[i](feats)
        feats = feats.view(feats.size(0), -1)
        return torch.softmax(self.fc(feats), dim=1)

# Load trained CAE safely
cae_model = CAEModel(num_classes=2).to(DEVICE)
state_dict = torch.load(CAE_PATH, map_location=DEVICE)
cae_model.load_state_dict(state_dict, strict=False)
cae_model.eval()
print("âœ… CAE model loaded and ready.")

# ==============================================================
# ðŸ§© Helper Functions
# ==============================================================
def auto_crop_board(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_green = np.array([25, 30, 30])
    upper_green = np.array([100, 255, 255])
    mask = cv2.inRange(hsv, lower_green, upper_green)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return image
    x, y, w, h = cv2.boundingRect(max(contours, key=cv2.contourArea))
    return image[y:y+h, x:x+w]

def extract_text(roi):
    if roi is None or roi.size == 0:
        return "unlabeled"
    roi = cv2.resize(roi, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    gray = cv2.fastNlMeansDenoising(gray, h=30)
    th = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                               cv2.THRESH_BINARY, 31, 2)
    kernel = np.ones((2,2), np.uint8)
    th = cv2.morphologyEx(th, cv2.MORPH_OPEN, kernel)
    txt = pytesseract.image_to_string(
        th,
        config="--psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
    ).strip().replace("\n", " ")
    return txt if txt else "unlabeled"

def preprocess_for_cae(roi):
    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
    ])
    return transform(roi).to(DEVICE)

def detect_with_yolo_ocr_cae(image):
    results = model(image)
    annotated = results[0].plot()
    comps = []

    for i, box in enumerate(results[0].boxes):
        cls_id = int(box.cls)
        label = results[0].names[cls_id]
        conf = float(box.conf.cpu().numpy())
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
        roi = image[y1:y2, x1:x2]
        ocr_text = extract_text(roi)

        # CAE inference
        try:
            roi_tensor = preprocess_for_cae(roi).unsqueeze(0)
            with torch.no_grad():
                cae_out = cae_model(roi_tensor)
            cae_conf = cae_out[0][1].item()  # probability of correct
        except Exception:
            cae_conf = 0.0

        if cae_conf > 0.75:
            final_label = f"{label} âœ…"
            color = (0, 255, 0)
        elif cae_conf > 0.5:
            final_label = f"{label}? ({ocr_text})"
            color = (0, 165, 255)
        else:
            final_label = f"{label} âŒ"
            color = (0, 0, 255)

        comps.append({
            "id": f"C{i}",
            "label": final_label,
            "ocr": ocr_text,
            "yolo_conf": round(conf, 3),
            "cae_conf": round(cae_conf, 3),
            "position": [x1, y1, x2 - x1, y2 - y1]
        })

        cv2.rectangle(annotated, (x1, y1), (x2, y2), color, 2)
        cv2.putText(annotated, final_label, (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.4, color, 1)

    return annotated, comps

# ==============================================================
# ðŸ“¤ Upload PCB Image
# ==============================================================
print("ðŸ“¸ Upload a single PCB image to test CAE integration")
uploaded = files.upload()
path = list(uploaded.keys())[0]
img = cv2.imread(path)

# ==============================================================
# ðŸ” Detection + Visualization (with NumPy to Python int fix)
# ==============================================================
crop = auto_crop_board(img)
detected, components = detect_with_yolo_ocr_cae(crop.copy())

# Convert NumPy int64 â†’ int for JSON serialization
for c in components:
    c["position"] = [int(x) for x in c["position"]]
    c["yolo_conf"] = float(c["yolo_conf"])
    c["cae_conf"] = float(c["cae_conf"])

# Save outputs
cv2.imwrite(f"{OUT_DIR}/detected_pcb_cae.png", detected)
with open(f"{OUT_DIR}/detected_components_cae.json", "w") as f:
    json.dump({"components": components}, f, indent=2)

print(f"\nâœ… Detection complete. Results saved to: {OUT_DIR}")
print("\nDetected components:")
for c in components:
    print(f" - {c['label']} | OCR: {c['ocr']} | YOLO Conf: {c['yolo_conf']} | CAE Conf: {c['cae_conf']}")

# ==============================================================
# ðŸ–¼ï¸ Display Annotated Image
# ==============================================================
plt.figure(figsize=(10,8))
plt.imshow(cv2.cvtColor(detected, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("PCB Detection (YOLO + Enhanced OCR + Correct CAE)")
plt.show()
```

```python
# ============================================================
# ðŸ§  YOLOv8 + Florence-2 Combined PCB Inspection Pipeline
# Loads trained model from Google Drive + fixes YOLO JSON export
# ============================================================

from ultralytics import YOLO
from transformers import AutoProcessor, AutoModelForCausalLM
from PIL import Image
import torch, json, os

# ------------------------------------------------------------
# STEP 1: Mount Google Drive & Load YOLO model
# ------------------------------------------------------------
from google.colab import drive
drive.mount('/content/drive')

# âœ… Use your verified model path
yolo_model_path = "/content/drive/MyDrive/YOLO/runs/pcb_yolov8_cap_res/weights/best.pt"

if not os.path.exists(yolo_model_path):
    raise FileNotFoundError(f"âŒ Model not found at: {yolo_model_path}")

print(f"âœ… YOLO weights loaded from: {yolo_model_path}")
yolo_model = YOLO(yolo_model_path)

# ------------------------------------------------------------
# STEP 2: Upload PCB image
# ------------------------------------------------------------
try:
    from google.colab import files
    print("\nðŸ“¸ Please upload your PCB image")
    uploaded = files.upload()
    image_path = list(uploaded.keys())[0]
    print(f"âœ… PCB image uploaded: {image_path}")
except Exception as e:
    image_path = input("Enter path to your PCB image: ")

# ------------------------------------------------------------
# STEP 3: Run YOLOv8 detection
# ------------------------------------------------------------
print("\nðŸ” Running YOLOv8 detection...")
yolo_results = yolo_model(image_path)
print("âœ… YOLO detection complete!")

# âœ… Convert YOLO results to JSON properly
yolo_json = yolo_results[0].to_json()  # <-- correct method
print("ðŸ§© YOLO JSON Export Complete!")

# Save annotated image
annotated_img = yolo_results[0].plot()
Image.fromarray(annotated_img).save("pcb_detected.jpg")
print("ðŸ–¼ï¸ Saved annotated image as pcb_detected.jpg")

# ------------------------------------------------------------
# STEP 4: Load Florence-2 model for captioning
# ------------------------------------------------------------
print("\nðŸ§  Loading Florence-2 model (this may take ~1 min)...")

model_id = "microsoft/Florence-2-large"
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype  = torch.float16 if device == "cuda" else torch.float32

processor = AutoProcessor.from_pretrained(model_id, trust_remote_code=True)
florence_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=dtype,
    device_map="auto" if device == "cuda" else None,
    trust_remote_code=True
).to(device)

# ------------------------------------------------------------
# STEP 5: Generate descriptive caption from Florence-2
# ------------------------------------------------------------
image = Image.open(image_path).convert("RGB")
prompt = "<CAPTION>"   # Florence captioning task token

inputs = processor(text=prompt, images=image, return_tensors="pt")
for k, v in inputs.items():
    if v.dtype.is_floating_point:
        inputs[k] = v.to(device=device, dtype=dtype)
    else:
        inputs[k] = v.to(device=device)

generated = florence_model.generate(**inputs, max_new_tokens=512)
florence_caption = processor.decode(generated[0], skip_special_tokens=True)

print("\nðŸ§¾ Florence-2 Description:\n", florence_caption)

# ------------------------------------------------------------
# STEP 6: Combine results into final inspection report
# ------------------------------------------------------------
try:
    detections = json.loads(yolo_json)
    component_summary = [
        f"- {d['name']} (confidence={d['confidence']:.2f})"
        for d in detections
    ]
except Exception as e:
    print("âš ï¸ Could not parse YOLO JSON:", e)
    component_summary = ["Could not parse YOLO output"]

report = f"""
========================================================
ðŸ§  PCB INSPECTION REPORT
========================================================
ðŸ“¸ Image: {image_path}

ðŸ“¦ Component Detections (YOLOv8)
--------------------------------
{chr(10).join(component_summary)}

ðŸ§© Florence-2 Visual Description
--------------------------------
{florence_caption}

ðŸ–¼ï¸ Annotated Image Saved:
pcb_detected.jpg

========================================================
"""

print(report)

with open("pcb_report.txt", "w") as f:
    f.write(report)
print("ðŸ’¾ Saved as pcb_report.txt in current directory")
```
